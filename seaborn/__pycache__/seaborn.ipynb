```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# loading dataset 
data = sns.load_dataset("iris") 
```

```python
plot = sns.FacetGrid(data, col="species") 
```

```python
plot.map(plt.plot, "sepal_width") 
```

```python
plt.show()
```

```python
data1 = sns.load_dataset("flights") 
```

```python
plot = sns.PairGrid(data1) 
plot.map(plt.plot) 
plt.show()
```

```python
sns.relplot(x='sepal_width', y='species', data=data) 
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=data) 
plt.show()
```

```python
sns.lineplot(x='sepal_length', y='species', data=data) 
plt.show()
```

```python
sns.barplot(x='species', y='sepal_length', data=data) 
plt.show()
```

```python
sns.countplot(x='species', data=data) 
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_width', data=data) 
plt.show()
```

```python
sns.violinplot(x='species', y='sepal_width', data=data) 
plt.show()
```

```python
sns.stripplot(x='species', y='sepal_width', data=data) 
plt.show()
```

```python
sns.swarmplot(x='species', y='sepal_width', data=data) 
plt.show()
```

```python
sns.histplot(x='species', y='sepal_width', data=data) 
plt.show()
```

```python
sns.distplot(data['sepal_width']) 
plt.show()
```

```python
sns.jointplot(x='species', y='sepal_width', data=data) 
plt.show()
```

```python
sns.pairplot(data=data, hue='species') 
plt.show()
```

```python
sns.rugplot(data=data) 
plt.show()
```

```python
sns.kdeplot(x='sepal_length', y='sepal_width', data=data) 
plt.show()
```

```python
data2 = sns.load_dataset("tips") 
  
sns.lmplot(x='total_bill', y='tip', data=data2) 
plt.show()
```
